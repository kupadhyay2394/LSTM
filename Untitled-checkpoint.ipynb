```python
df=pd.read_csv('AAPl.csv')
```

```python
df.head()
```

```python
#to add index
df1=df.reset_index()['close']
plt.plot(df1)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler=MinMaxScaler(feature_range=(0,1))
df1=scaler.fit_transform(nm.array(df1).reshape(-1,1))
```

```python
data_training=int(len(df1)*0.65)
data_testing=len(df1)-data_training
train_Data, test_Data=df1[0:data_training,:],df1[data_training:len(df1):1]
```

```python
def create_dataset(dataset,timestep=1):
    dataX, dataY=[],[]
    for i in range(len(dataset)-timestep-1):
        a=dataset[i:(i+timestep),0]
        dataX.append(a)
        dataY.append(dataset[(timestep+i),0])
    return nm.array(dataX), nm.array(dataY)   
```

```python
timestep=100
trainX,trainY=create_dataset(train_Data,timestep)
testX,testY=create_dataset(test_Data,timestep)
```

```python
print(trainX.shape), print(trainY.shape)
```

```python
trainX =trainX.reshape(trainX.shape[0],trainX.shape[1],1)
testX =testX.reshape(testX.shape[0],testX.shape[1],1)
```

```python
testX.shape,testY.shape
```

```python
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.models import Sequential 
```

```python
model=Sequential()
model.add(LSTM(50, return_sequences=True,input_shape=(100,1)))


model.add(LSTM(50, return_sequences=True))


model.add(LSTM(50))




model.add(Dense(1))
model.compile(optimizer='adam',loss='mean_squared_error') 
    
```

```python
model.summary()
```

```python
model.fit(trainX,trainY,validation_data=(testX,testY), epochs=100,verbose=1,batch_size=64)
```

```python
import tensorflow as tf
```

```python
train_predict=model.predict(trainX)
test_predict=model.predict(testX)
```

```python
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```
